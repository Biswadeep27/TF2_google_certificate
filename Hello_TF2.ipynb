# The Hello World of Neural Network

In this example we will try to fit in a model for a simple problem of following linear regression.

$$
y = 2X -1 
$$

```
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

# Define and compile the neural Network



*   we will create a neural netowork with 1 layer which has 1 neuron and the input shape to this is just 1 value.
*   **Optimizer** : to guess the linear regression function : **Stochastic gradient descent**.
*   **loss** : to calculate the deviation between predicted output and correct known output : **Mean squared Error**
*   Based on how the loss function went, the Optimizer will try to minimize the loss at every next **Epoch**.





```
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
```

# Providing the Data

```
xs = np.array([-1,0,1,2,3,4,5,6,7],dtype=float)
ys = np.array([-3,-1,1,3,5,7,9,11,13],dtype=float)
```

# Training the neural network

```
model.fit(xs,ys,epochs=500)
```

```
print(model.predict([20.0]))
```
