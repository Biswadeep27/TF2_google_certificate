```python
ele = [1,2,3,4,8,10,12,20]

res = []
i = 0
while i<len(ele)-1:
    if ele[i] + 1 != ele[i+1]:
        diff = ele[i+1] - ele[i]
        for j in range(diff-1):
            #print(j+1)
            res.append(ele[i]+j+1)
    i += 1

print(res)
```
